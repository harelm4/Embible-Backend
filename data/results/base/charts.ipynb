```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
fine_tuned=pd.read_csv('fine_tuned.csv')
no_fine_tuned=pd.read_csv('no_fine_tuned.csv')
fine_tuned
```

```python
no_fine_tuned['model']=no_fine_tuned['model'].replace('tau/tavbert-he','tavBert')
no_fine_tuned['model']=no_fine_tuned['model'].replace('distilbert-base-uncased','distilbert')
no_fine_tuned['model']=no_fine_tuned['model'].replace('bert-base-multilingual-cased','mBert')
no_fine_tuned['model']=no_fine_tuned['model'].replace('dicta-il/alephbertgimmel-base','aleph-bert-gimmel')
no_fine_tuned
```

```python
fine_tuned['model']=fine_tuned['model'].replace('Embible/tavbert-20-epochs','tavBert')
fine_tuned['model']=fine_tuned['model'].replace('Embible/distilBert-20-epochs','distilbert')
fine_tuned['model']=fine_tuned['model'].replace('Embible/mBert-20-epochs','mBert')
fine_tuned['model']=fine_tuned['model'].replace('Embible/AlephBertGimmel-20-epochs','aleph-bert-gimmel')
fine_tuned
```

```python
def replace_file(df):
    for mask_precent in [5,10,15]:
        df['file']=df['file'].replace(f'../../data/Hit@K/masked MIX char tokens/mix_{mask_precent}.json',mask_precent)

replace_file(fine_tuned)
replace_file(no_fine_tuned)
    
```

```python
fine_tuned
```

```python
def plot(df,title):
    df = df.sort_values('hit@k', ascending=False)
    sns.barplot(x='model',y='hit@k',hue='file',data=df)
    plt.legend().set_title(None)
    plt.ylabel().set_title('Precision (Hit@1)')

    plt.title(title)
    plt.show()

plot(fine_tuned,'Fine Tuned Models')
plot(no_fine_tuned,'As Is model')
```
