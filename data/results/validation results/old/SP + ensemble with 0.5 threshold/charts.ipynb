```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
titles = ['CharHitAt1','CharHitAt5','WordHitAt1','WordHitAt5','MRR_CHAR','MRR_WORD']
dfs={title:pd.read_csv(title+'.csv') for title in titles}
list(dfs.items())[0][1]['file'].tolist()
```

```python
for k,df in dfs.items():
    df['model']=df['model'].replace('EnsembleV2_iterative_sp','Iterative')
    df['model']=df['model'].replace('EnsembleV2_recurrent_sp','Recursive')
    df['model']=df['model'].replace('EnsembleV2_baseline_sp','Baseline')
    for p in [5,10,15]:
        df['file']=df['file'].replace(f'../../data/Hit@K/masked MIX char tokens/mix_{p}.json',f'{p}% mask')

list(dfs.items())[0][1]

```

```python
for k,df in dfs.items():
    sns.barplot(data=df,x='file',y='hit@k',hue='model')
    plt.xlabel('')
    plt.ylabel('')
    plt.title(k)
    plt.show()
```
